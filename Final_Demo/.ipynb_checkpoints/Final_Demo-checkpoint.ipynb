### 2. To a dictionary containing each sign's average (prop_all) for each emotion, append each sign's average

```python
prop_all = {'positive': [], 'negative': [], 'fear': [], 'anger': [], 'anticip': [], 'trust': [], 'surprise': [],
          'sadness': [], 'disgust': [], 'joy': []}

for emotion in libra_averages[3]:
    if emotion == "anticipation":
            prop_all['anticip'].append(libra_averages[3][emotion])
    else:
        prop_all[emotion].append(libra_averages[3][emotion])

for emotion in gemini_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(gemini_averages[3][emotion])
    else:
        prop_all[emotion].append(gemini_averages[3][emotion])

for emotion in leo_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(leo_averages[3][emotion])
    else:
        prop_all[emotion].append(leo_averages[3][emotion])
        
for emotion in aries_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(aries_averages[3][emotion])
    else:
        prop_all[emotion].append(aries_averages[3][emotion])
        
for emotion in taurus_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(taurus_averages[3][emotion])
    else:
        prop_all[emotion].append(taurus_averages[3][emotion])

for emotion in cancer_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(cancer_averages[3][emotion])
    else:
        prop_all[emotion].append(cancer_averages[3][emotion])
        
for emotion in virgo_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(virgo_averages[3][emotion])
    else:
        prop_all[emotion].append(virgo_averages[3][emotion])
        
for emotion in scorp_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(scorp_averages[3][emotion])
    else:
        prop_all[emotion].append(scorp_averages[3][emotion])
        
for emotion in sag_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(sag_averages[3][emotion])
    else:
        prop_all[emotion].append(sag_averages[3][emotion])
        
for emotion in cap_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(cap_averages[3][emotion])
    else:
        prop_all[emotion].append(cap_averages[3][emotion])
        
for emotion in aq_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(aq_averages[3][emotion])
    else:
        prop_all[emotion].append(aq_averages[3][emotion])
        
for emotion in pisces_averages[3]:
    if emotion == "anticipation":
        prop_all['anticip'].append(pisces_averages[3][emotion])
    else:
        prop_all[emotion].append(pisces_averages[3][emotion])
        
prop_all
```

## NRCLex

### 3. Enter the input text of the tweet you want to guess the sign of

```python
input = "Hate on me, I blew but I'm the same OG People warned me: 'When you're on top there's envy'"
```

### 4. Create NRCLex object of the input text, perform emotional analysis, and store proportions/emotion profile in dictionary (prop_input)

```python
# ----------------------------------------------------------------------------------------

# NRCLex

# ----------------------------------------------------------------------------------------

# Create text object
text_object = NRCLex(input)

# Store emotion score frequency (averages of emotion words, part / whole)
input_emotions_freq = text_object.affect_frequencies

# Store frequencies in dictionary 
prop_input = dict()
for emotion in input_emotions_freq:
    if emotion == "anticipation":
        emotion = "anticip"
    prop_input[emotion] = input_emotions_freq[emotion]

input_emotions_freq
```

### 5. Apply a lambda function to get the "shortest distance" average from each emotion, and store in results

```python
# Get the number that is closest to our input number for each emotion, and append to results dictionary
results = []
for emotion in prop_input:
    
    if prop_input[emotion] != 0:
        
    
        # Lambda function = function without a name. Iterates through each emotion and finds the minimum absolute value from
        # the input average of that emotion.
        output = min(prop_all[emotion], key=lambda x: abs(x-prop_input[emotion]))

        results.append({emotion: output})
    else:
        results.append({emotion: 0})

results
```

### 6. How many of the closest averages are the averages of each sign? Get a count for each sign

```python
# Count how many closest numbers are libra, gemini, or leo. Replace with name and number.
libra_count = 0
gem_count = 0
leo_count = 0
aries_count = 0
taurus_count = 0
cancer_count = 0
virgo_count = 0
scorp_count = 0
sag_count = 0
cap_count = 0
aq_count = 0
pisces_count = 0

counts = []

for i in results:
    for emotion in i:
        if i[emotion] != 0:
            if i[emotion] == prop_all[emotion][0]:
                i[emotion] = ['LIBRA', prop_all[emotion][0]]
                libra_count += 1

            elif i[emotion] == prop_all[emotion][1]:
                i[emotion] = ['GEMINI', prop_all[emotion][1]]
                gem_count += 1

            elif i[emotion] == prop_all[emotion][2]:
                i[emotion] = ['LEO', prop_all[emotion][2]]
                leo_count += 1

            elif i[emotion] == prop_all[emotion][3]:
                i[emotion] = ['ARIES', prop_all[emotion][3]]
                aries_count += 1

            elif i[emotion] == prop_all[emotion][4]:
                i[emotion] = ['TAURUS', prop_all[emotion][4]]
                taurus_count += 1

            elif i[emotion] == prop_all[emotion][5]:
                i[emotion] = ['CANCER', prop_all[emotion][5]]
                cancer_count += 1

            elif i[emotion] == prop_all[emotion][6]:
                i[emotion] = ['VIRGO', prop_all[emotion][6]]
                virgo_count += 1

            elif i[emotion] == prop_all[emotion][7]:
                i[emotion] = ['SCORPIO', prop_all[emotion][7]]
                scorp_count += 1

            elif i[emotion] == prop_all[emotion][8]:
                i[emotion] = ['SAGITTARIUS', prop_all[emotion][8]]
                sag_count += 1

            elif i[emotion] == prop_all[emotion][9]:
                i[emotion] = ['CAPRICORN', prop_all[emotion][9]]
                cap_count += 1

            elif i[emotion] == prop_all[emotion][10]:
                i[emotion] = ['AQUARIUS', prop_all[emotion][10]]
                aq_count += 1

            elif i[emotion] == prop_all[emotion][11]:
                i[emotion] = ['PISCES', prop_all[emotion][11]]
                pisces_count += 1

counts.append(libra_count)
counts.append(gem_count)
counts.append(leo_count)
counts.append(aries_count)
counts.append(taurus_count)
counts.append(cancer_count)
counts.append(virgo_count)
counts.append(scorp_count)
counts.append(sag_count)
counts.append(cap_count)
counts.append(aq_count)
counts.append(pisces_count)

counts
```

## Output

### 7. Depending on the position of the max count sign, we determine our final prediction.

```python
# Get the index of the max number of matches
final_verdict = counts.index(max(counts))

# Match to a star sign (0 = Libra, 1 = Gemini, 2 = Leo) based on how we appended

print("PREDICTION:")

if final_verdict == 0:
    print("LIBRA")
elif final_verdict == 1:
    print("GEMINI")
elif final_verdict == 2:
    print("LEO")
elif final_verdict == 3:
    print("ARIES")
elif final_verdict == 4:
    print("TAURUS")
elif final_verdict == 5:
    print("CANCER")
elif final_verdict == 6:
    print("VIRGO")
elif final_verdict == 7:
    print("SCORPIO")
elif final_verdict == 8:
    print("SAGITTARIUS")
elif final_verdict == 9:
    print("CAPRICORN")
elif final_verdict == 10:
    print("AQUARIUS")
elif final_verdict == 11:
    print("PISCES")

    
#print()

# Frequencies for input string
#print("INPUT AVERAGES: " + str(prop_input))
#print()
# Sign frequencies per emotion, [0] = Libra, [1] = Gem, [2] = Leo
#print("SIGN AVERAGES: " + str(prop_all))# json.dumps(prop_all, indent=4))
#print()
# Output pretty with indents
#print(json.dumps(results,indent=4))

# 
#Example:

#'fear' input = .125
#fear for all signs = [0.07434873958991314, 0.04, 0.3]
#                        Libra               G    Leo 
#*Libra (.074) is the closest to .125*
#
```
